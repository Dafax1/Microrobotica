```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from PIL import Image
from joblib import dump, load
```

```python
data = pd.read_csv(".\data.csv", na_values=[-99999.0])
data.dropna(inplace=True)
data
```

```python
year = data["Year"].values
sales = data["Sales of cigarettes per adult per day"].values
rate_male = data["age-standardized_death_rate_per_100k_male"].values
rate_female = data["age-standardized_death_rate_per_100k_female"].values
```

```python
year
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(year, 10*sales, label="Vendite di sigarette pro capite * 10")
ax.plot(year, rate_male, label="Tasso di mortalitÃ  per i maschi")
ax.plot(year, rate_female, label="Tasso di mortalitÃ  per le femmine")

ax.set_xlabel("Anno")
ax.set_ylabel("Freq")
ax.grid()
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(sales, rate_male, label="Tasso di mortalitÃ  per i maschi")

ax.set_xlabel("Vendita di siga al giorno per persona")
ax.set_ylabel("Tasso mortalitÃ  maschile")

ax.grid()
plt.show()
```

```python
#facciamo tante regressioni e troviamo quelle con l'r^2 maggiore (troviamo piÃ¹ lag)
```

```python
for lag in range (5,30):
    rate_male_lagged = np.roll(rate_male, -lag)[:-lag]
    
    regressione = LinearRegression()
    regressione.fit(rate_male_lagged[:, np.newaxis], sales[:-lag])
```

```python


fig, ax = plt.subplots(figsize=(6,6))
ax.plot(sales, rate_male_lagged, label="Tasso di mortalitÃ  per i maschi")
ax.plot(sales[:-lag], rate_male_lagged, label="Tasso di mortalitÃ  per i maschi")

ax.set_xlabel("Vendita di siga al giorno per persona")
ax.set_ylabel("Tasso mortalitÃ  maschile")

ax.grid()
plt.show()
```
